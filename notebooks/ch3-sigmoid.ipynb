```python
import torch
import matplotlib.pyplot as plt

# 创建输入张量 x，范围从 -10 到 10，共 200 个点
x = torch.linspace(-10, 10, steps=200)

# 计算 y = sigmoid(x)
y = torch.sigmoid(x)

# 转为 NumPy 以便绘图（因为 matplotlib 不直接支持 torch.Tensor）
x_np = x.numpy()
y_np = y.numpy()

# 绘制 sigmoid 曲线
plt.figure(figsize=(8, 5))
plt.plot(x_np, y_np, label=r'$y = \sigma(x) = \frac{1}{1 + e^{-x}}$', color='blue')
plt.axhline(y=0.5, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(x=0, color='gray', linestyle='--', linewidth=0.8)
plt.title('Sigmoid')
plt.xlabel('x')
plt.ylabel('y')
```
