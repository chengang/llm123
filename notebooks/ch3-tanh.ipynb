```python
import torch
import matplotlib.pyplot as plt

# 创建输入张量 x，范围从 -10 到 10，共 200 个点
x = torch.linspace(-10, 10, steps=200)

# 计算 y = tanh(x)
y = torch.tanh(x)

# 转为 NumPy 以便绘图（因为 matplotlib 不直接支持 torch.Tensor）
x_np = x.numpy()
y_np = y.numpy()

# 绘制 tanh 曲线
plt.figure(figsize=(8, 5))
plt.plot(x_np, y_np)
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.8) # 中心线
plt.axvline(x=0, color='gray', linestyle='--', linewidth=0.8) # 中心线
plt.title('Tanh')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
