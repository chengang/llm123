```python
import torch
import matplotlib.pyplot as plt

# 生成 x 值：从 -5 到 5（ReLU 在 x > 0 时为线性增长，在 x < 0 时输出为 0）
x = torch.linspace(-5, 5, steps=200)
y = torch.relu(x)

# 转为 NumPy 用于绘图
plt.figure(figsize=(8, 5))
plt.plot(x.numpy(), y.numpy(), color='blue', linewidth=2)

# 添加参考线
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(x=0, color='gray', linestyle='--', linewidth=0.8)

# 设置标题和标签
plt.title('ReLU')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 5)
plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()
```
